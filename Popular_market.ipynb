### 전국에서 가장 인기있는 업종은?

```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv', encoding='euc-kr')
```

```python
print(df.head())  # 데이터프레임이 잘 만들어졌는지 확인
```

```python
# 전국 100대 업종 추출
job_category = df['업종'].unique()
job_category
```

```python
for i in job_category:
    employer_sum = df.groupby('업종')['당월'].sum()
employer_sum
```

```python
current_result = pd.DataFrame(employer_sum).reset_index()
current_result
```

```python
sorted_current = current_result.sort_values(by='당월', ascending=False)
sorted_current
```

```python
top30_current = sorted_current.head(30)
top30_current
```

```python
# 작년의 업종별 사업자 수
for i in job_category:
    employer_lastYear = df.groupby('업종')['전년동월'].sum()
employer_lastYear
```

```python
lastYear_result = pd.DataFrame(employer_lastYear).reset_index()
lastYear_result
```

```python
sorted_lastYear = lastYear_result.sort_values(by='전년동월', ascending=False)
sorted_lastYear
```

```python
top30_lastYear = sorted_lastYear.head(30)
top30_lastYear
```

```python
top30_current.to_csv('top30_current.csv', index=False)
top30_lastYear.to_csv('top30_lastYear.csv', index=False)
```

```python
merged_df = pd.merge(top30_current, top30_lastYear, on='업종', how='inner')
merged_df
```

```python
concatenated_df = pd.concat([top30_current, top30_lastYear], axis=1)  # Concatenate along columns
concatenated_df
```

#### 1~27위는 작년과 차이가 없지만, 28위부터 순위 변동이 있다는 점을 확인

```python
middle_current = sorted_current.iloc[27:]
middle_current
```

```python
middle_lastYear = sorted_lastYear.iloc[27:]
middle_lastYear
```

```python
merged_df.to_csv('merged.csv', index=False)
```

```python
for i in job_category:
    region = df.groupby(['업종', '시도'])['당월'].sum()
region_df = pd.DataFrame(region)
region_df
```

```python
region_df.to_csv("region.csv")
```

##### 이제 지역 별 인기 업종 순위를 알아봅시다. 

```python
regions = df['시도'].unique()
regions
```

```python
for i in regions:
    region_sum = df.groupby('시도')['당월'].sum()
region_sum
```

```python
for i in regions:
    region_sum_sf = df.groupby(['시도', '시군구'])['당월'].sum()
region_sum_sf
```

```python
for i in regions:
    region_sum_job = df.groupby(['시도', '시군구', '업종'])['당월'].sum()
region_sum_job
```

```python
region_sum_df = pd.DataFrame(region_sum)
region_sum_sf_df = pd.DataFrame(region_sum_sf)
region_sum_job_df = pd.DataFrame(region_sum_job)
```

```python
region_sum_df.to_csv("business_sum_region.csv")
region_sum_sf_df.to_csv("business_sum_region_specific.csv")
region_sum_job_df.to_csv("business_sum_by_region.csv")
```
